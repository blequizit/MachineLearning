# **Dicionário de dados**

Car_ID			:	Unique id of each observation (Interger)

Symboling 			:	Its assigned insurance risk rating, A value of +3 indicates that the auto is risky, -3 that it is probably
pretty safe.(Categorical)

carCompany			:	Name of car company (Categorical)

fueltype			:	Car fuel type i.e gas or diesel (Categorical)

aspiration			:	Aspiration used in a car (Categorical)

doornumber			:	Number of doors in a car (Categorical)

carbody			:	body of car (Categorical)

drivewheel			:	type of drive wheel (Categorical)

enginelocation			:	Location of car engine (Categorical)

wheelbase			:	Weelbase of car (Numeric)

carlength			:	Length of car (Numeric)

carwidth			:	Width of car (Numeric)

carheight			:	height of car (Numeric)

curbweight			:	The weight of a car without occupants or baggage. (Numeric)

enginetype			:	Type of engine. (Categorical)

cylindernumber			:	cylinder placed in the car (Categorical)

enginesize			:	Size of car (Numeric)

fuelsystem			:	Fuel system of car (Categorical)

boreratio			:	Boreratio of car (Numeric)

stroke			:	Stroke or volume inside the engine (Numeric)

compressionratio			:	compression ratio of car (Numeric)

horsepower			:	Horsepower (Numeric)

peakrpm			:	car peak rpm (Numeric)

citympg			:	Mileage in city (Numeric)

highwaympg			:	Mileage on highway (Numeric)

price(Dependent variable)			:	Price of car (Numeric)

## Exploração Grafica

#### Distribuição da variável alvo (Price)

```python
fig = px.histogram(df, x='price', nbins=30, title='Distribuição dos preços dos carros', color_discrete_sequence=['blue'])
fig.show()
```

##### Distribuição das variáveis categóricas

```python
cat_cols = df.select_dtypes(include='object') # Separa somente as colunas com data type object

for col in cat_cols:
    fig = px.histogram(df, x=col, title=f'Distribuição das {col}', color_discrete_sequence=['purple'])
    fig.show()
```

#### Correlação entre variáveis numéricas

```python
df_num = df.select_dtypes(exclude='object') # Gera dataset com apenas variáveis numericas
correlation = df_num.corr() # gera matriz de correlação
fig = px.imshow(correlation,title='Mapa de calor das correlações do dataset de vendas de carros')
fig.show()
```

##### Exploração Price

```python
fig = px.scatter(df, x="enginesize", y="price", trendline="ols",
                 title="Relação entre o tamanho do motor e o preço do carro")
fig.show()
```

```python
fig = px.box(df, x="carbody", y="price", color="carbody", title="Distribuição de preços por tipo de carroceria")
fig.show()
```

```python
fig = px.box(df, x="fueltype", y="price", color="fueltype", title="Distribuição de preços por tipo de combustivel")
fig.show()
```

```python
fig = px.scatter(df, x="curbweight", y="price", trendline="ols",
                 title="Relação entre o peso do carro e o preço do carro")
fig.show()
```

```python
# Valor médio de venda dos carros por tipo de combustivel
fig = px.bar(df.groupby("fueltype")["price"].mean().reset_index(), x="fueltype", y="price",
              title="Valor médio de venda dos carros por tipo de combustível", color_discrete_sequence=["orange"])
fig.show()
```

```python
fig = px.box(df, x="aspiration", y="price", color="aspiration", title="Distribuição de preços por tipo aspiração do motor")
fig.show()
```

```python
# Valor médio de venda dos carros por tipo de combustivel
fig = px.bar(df.groupby("aspiration")["price"].mean().reset_index(), x="aspiration", y="price",
              title="Valor médio de venda dos carros por tipo de Aspiração", color_discrete_sequence=["orange"])
fig.show()
```

```python
fig = px.box(df, x="drivewheel", y="price", color="drivewheel", title="Distribuição de preços por tipo de tração")
fig.show()
```

```python
fig = px.box(df, x="cylindernumber", y="price", color="cylindernumber", title="Distribuição de preços por Quantidade de cilindros no motor")
fig.show()
```
